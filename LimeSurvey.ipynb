```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

c0 = ['Άλλο / Δεν επιθυμώ να απαντήσω', 'Άνδρας', 'Γυναίκα']

c1 = ['18-27', '28-36', '37-45', '46-54', '55-64', '>= 65']

c2 = ['Άνεργος/η', 'Δημόσιος/α Υπάλληλος',
       'Ελεύθερος/η Επαγγελματίας – Αυτοαπασχολούμενος/η',
       'Ιδιωτικός/ή Υπάλληλος', 'Μαθητής/τρια – Φοιτητής/τρια', 'Οικιακά',
       'Συνταξιούχος']
       
c3 = ['Ανώτατη εκπαίδευση', 'Απόφοιτος/η Γυμνασίου', 'Απόφοιτος/η Δημοτικού',
       'Απόφοιτος/η Λυκείου',
       'Κάτοχος μεταπτυχιακού / διδακτορικού τίτλου',
       'Μεταλυκειακή εκπαίδευση']
       
c4 = ['12.001€ – 24.000€', '24.001€ – 36.000€', '7.000€ – 12.000€',
       '< 7.000€', '> 36.000€', 'Δεν επιθυμώ να απαντήσω']

c5 = ['Δήμος Αμπελοκήπων – Μενεμένης', 'Δήμος Βόλβης', 'Δήμος Δέλτα',
       'Δήμος Θέρμης', 'Δήμος Θερμαϊκού', 'Δήμος Θεσσαλονίκης',
       'Δήμος Καλαμαριάς', 'Δήμος Κορδελιού – Ευόσμου', 'Δήμος Λαγκαδά',
       'Δήμος Νεάπολης – Συκεών', 'Δήμος Παύλου Μελά',
       'Δήμος Πυλαίας – Χορτιάτη', 'Δήμος Χαλκηδόνος',
       'Δήμος Ωραιοκάστρου']

c6 = ['Όχι, δεν έχω πρόσβαση',
       'Δεν είμαι κάτοχος, αλλά έχω πρόσβαση σε Ι.Χ. όχημα με το οποίο μετακινούμαι',
       'Ναι, είμαι κάτοχος Ι.Χ. οχήματος']
       
c7 = ['Ηλεκτρικό Πατίνι', 'Ι.Χ. όχημα', 'Λεωφορείο',
       'Μοτοσικλέτα / Μοτοποδήλατο', 'Περπάτημα', 'Ποδήλατο', 'Ταξί']

c8 = ['1-2 ημέρες / εβδομάδα', '1-3 ημέρες / μήνα',
       '1-5 ημέρες / εξάμηνο', '3-5 ημέρες / εβδομάδα',
       '6-7 ημέρες / εβδομάδα', 'Καθόλου',
       'Λιγότερο από 1 ημέρα / εξάμηνο']
       
c9 = ['Εργασία', 'Εκπαίδευση', 'Αγορές ειδών πρώτης ανάγκης (σουπερμάρκετ, φαρμακείο, κλπ.)', 
        'Λοιπές αγορές', 'Αναψυχή / Ψυχαγωγία', 
        'Ατομικές ή οικογενειακές υποθέσεις (ιατρικά ραντεβού, συναλλαγές με τράπεζες / δημόσιο, κλπ.)', 'Άλλο']
        
c10 = ['Αρκετά', 'Καθόλου', 'Λίγο', 'Μέτρια', 'Πολύ']

c11 = ['Μερικές φορές', 'Πολύ Συχνά', 'Ποτέ', 'Σπάνια', 'Συχνά']

c12 = ['Κάπως λιγότερο', 'Κάπως περισσότερο', 'Περίπου το ίδιο',
       'Πολύ λιγότερο', 'Πολύ περισσότερο']
       
c13 = ['Δεν επιθυμώ να απαντήσω', 'Δεν θα έχω εργασία',
       "Θα δουλεύω αποκλειστικά εξ 'αποστάσεως (τηλεργασία)",
       'Θα δουλεύω αποκλειστικά με φυσική παρουσία στον εργασιακό χώρο',
       "Θα δουλεύω εκ περιτροπής είτε εξ 'αποστάσεως είτε με φυσική παρουσία στον εργασιακό χώρο", 'Άλλο']

c14 = ['Δουλεύω αποκλειστικά εξ’ αποστάσεως (τηλεργασία)', 'Δουλεύω εκ περιτροπής είτε εξ ’αποστάσεως είτε με φυσική παρουσία στον εργασιακό χώρο', 
      'Δουλεύω αποκλειστικά με φυσική παρουσία στον εργασιακό χώρο', 'Είμαι σε καθεστώς προσωρινής αναστολής εργασίας',
       "Είμαι σε άδεια άνευ ή μετ' αποδοχών", 'Άλλο']

for i in range(15):
    vars()['c' + str(i)] = dict(zip(vars()['c' + str(i)], np.arange(len(vars()['c' + str(i)]))))

ans = [c0, c1, c2, c3, c4, c5, c6, c7, c8, c9, c10, c11, c12, c13, c14]

df = pd.read_csv('results-survey144827-1.csv', index_col = 0)
df = df[df['Σε ποιο Δήμο διαμένετε;'] != 'Άλλο']
df = df.drop(['Ημερομηνία καταχώρισης', 'Τελευταία σελίδα', 'Αρχική γλώσσα', 'Seed', 'Σε ποιο Δήμο διαμένετε; [Άλλο]'], 1)

data = df.copy()

def merg_cond(par1, par2, df):
    df[par1] = np.where((df[par1]=='Άλλο'), df[par2], df[par1])
    df[par1] = df[par1].astype(str)
    df = df.drop([par2], 1)
    return df

df = merg_cond('Ποιος ήταν ο σκοπός της κυριότερης μετακίνησης σας;', 'Ποιος ήταν ο σκοπός της κυριότερης μετακίνησης σας; [Άλλο]', df)
df = merg_cond('Ποιά είναι η εργασιακή σας κατάσταση κατά τη διάρκεια της πανδημίας;', 'Ποιά είναι η εργασιακή σας κατάσταση κατά τη διάρκεια της πανδημίας; [Άλλο]', df)
df = merg_cond('Στο μέλλον , ποιό από τα παρακάτω πιστεύετε ότι θα περιγράφει καλύτερα την εργασιακή σας κατάσταση; ', 'Στο μέλλον , ποιό από τα παρακάτω πιστεύετε ότι θα περιγράφει καλύτερα την εργασιακή σας κατάσταση;  [Άλλο]', df)

data = df.copy()

df.iloc[:, 0] = (df.iloc[:, 0]).replace(ans[0])
df.iloc[:, 1] = (df.iloc[:, 1]).replace(ans[1])
df.iloc[:, 2] = (df.iloc[:, 2]).replace(ans[2])
df.iloc[:, 3] = (df.iloc[:, 3]).replace(ans[3])
df.iloc[:, 4] = (df.iloc[:, 4]).replace(ans[4])
df.iloc[:, 5] = (df.iloc[:, 5]).replace(ans[5])
df.iloc[:, 6] = (df.iloc[:, 6]).replace(ans[6])
df.iloc[:, 7] = (df.iloc[:, 7]).replace(ans[7])
df.iloc[:, 8] = (df.iloc[:, 8]).replace(ans[8])
df.iloc[:, 9] = (df.iloc[:, 9]).replace(ans[9]).apply(lambda x: pd.to_numeric(x, errors='coerce')).fillna(6)
df.iloc[:, 10:20] = (df.iloc[:, 10:20]).replace(ans[10])
df.iloc[:, 20:28] = (df.iloc[:, 20:28]).replace(ans[11])
df.iloc[:, 28] = (df.iloc[:, 28]).replace(ans[7])
df.iloc[:, 29] = (df.iloc[:, 29]).replace(ans[8])
df.iloc[:, 30:65] = (df.iloc[:, 30:65]).replace(ans[10])
df.iloc[:, 65] = (df.iloc[:, 65]).replace(ans[14]).apply(lambda x: pd.to_numeric(x, errors='coerce')).fillna(5)
df.iloc[:, 66] = (df.iloc[:, 66]).replace(ans[12])
df.iloc[:, 67] = (df.iloc[:, 67]).replace(ans[13]).apply(lambda x: pd.to_numeric(x, errors='coerce')).fillna(5)
df.iloc[:, 68:] = (df.iloc[:, 68:]).replace(ans[10])

df = df.astype(int)
df_exp = df.copy()
cols = np.arange((len(df_exp.columns)))
df_exp.columns = cols
df_exp = df_exp.add_prefix('Q')

df_exp['hue'] = data.iloc[:, 0] #allazei to hue
df_exp
```
